```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Train_Dataset-01.csv")
df
```

```python
df.isnull().sum()
df.isna().sum()
#missing values present
```

```python
from sklearn.impute import SimpleImputer
df = df.replace('',np.NaN)
df
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_Y = LabelEncoder()
df['diagnosis'] = labelencoder_Y.fit_transform(df['diagnosis'])
```

```python
x_train=df.loc[:, df.columns != 'diagnosis']
#x_train=df[['texture_worst','concavity_worst']]
y_train=df['diagnosis']
y_train
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer = imputer.fit(x_train)
x_train = imputer.transform(x_train)
```

```python
x_test=pd.read_csv('Test_Dataset-01.csv')
#x_test=x_test[['texture_worst','concavity_worst']]
```

```python
#normalization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0, max_depth=3)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
#Training 
print("Training Score:")
classifier.score(x_train,y_train)
```

```python
y_pred
#predicted values for the test dataset
```

```python
from sklearn.tree import export_graphviz
export_graphviz(classifier,  out_file='tree.dot')
```

```python
df1 = pd.read_csv("output.csv")
df1["decisionTree_pred"] = y_pred
df1.to_csv("output.csv", index=False)
```
